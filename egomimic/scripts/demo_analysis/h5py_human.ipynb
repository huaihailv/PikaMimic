### iteration on dir  

```python
human_data_path = "/home/lvhuaihai/data_groceries_video"

import os 

def find_episode_files(root_dir):
        # 用于存储匹配文件的路径
        episode_files = []

        # 遍历根目录下的所有文件和文件夹
        for dirpath, dirnames, filenames in os.walk(root_dir):
            # 遍历当前文件夹中的每个文件
            for filename in filenames:
                # 检查文件名是否匹配 'episode_xx.hdf5' 模式
                if filename.startswith('episode_'):
                    # 获取文件的绝对路径
                    file_path = os.path.join(dirpath, filename)
                    episode_files.append(file_path)
        return episode_files
    
files_list = find_episode_files(human_data_path)
```

```python
for i in range(len(files_list)):
    print("start to handle file: " + files_list[i])
    human_data: h5py.File = h5py.File(files_list[i], 'r+')
    head_pose_array = human_data['head_pose']
    head_pose_array = np.array(head_pose_array)
    non_zero_indices = np.nonzero(head_pose_array.any(axis=1))[0]
    actual_length = non_zero_indices[-1] + 1
    for key in human_data.keys():
        new_data = human_data[key][:actual_length]
        del human_data[key]
        human_data.create_dataset(key, data=new_data, shape=new_data.shape, dtype=new_data.dtype)
    human_data.close()
```
